# Afficher une courbe cumullé en fonction du temps 

```python
import pandas as pd

def Cummuler_Journalier(tab, colonne=None, valeurs=None):
    """
    Cumul progressif (jour après jour) de Revenu / Dépense / Ecart.
    Peut être global ou regroupé par valeur d'une colonne.

    Paramètres
    ----------
    tab : pd.DataFrame
        Doit contenir les colonnes ["Date", "Type", "Valeur"]
    colonne : str ou None
        Nom de colonne pour regrouper (ex: "Categorie").
        Si None → calcule seulement le global.
    valeurs : list ou None
        Liste des valeurs à prendre en compte (si colonne est spécifiée).
        Si None → toutes les valeurs uniques.

    Retour
    ------
    pd.DataFrame avec ["Date", "Revenu", "Depense", "Ecart"] (+ colonne si spécifiée)
    """

    tab["Type"] = tab["Type"].str.lower()
    tab["Date"] = pd.to_datetime(tab["Date"])

    date_min = tab["Date"].dt.date.min()
    date_max = tab["Date"].dt.date.max()
    all_days = pd.date_range(date_min, date_max, freq="D")

    resultats = []

    if colonne is None:
        cumul_revenu = 0
        cumul_depense = 0
        cumul_ecart = 0

        for d in all_days:
            sous_tab = tab[tab["Date"].dt.date == d.date()]
            revenu = sous_tab[sous_tab["Type"] == "revenu"]["Valeur"].sum()
            depense = sous_tab[sous_tab["Type"] == "depense"]["Valeur"].sum()

            cumul_revenu += revenu
            cumul_depense += depense
            cumul_ecart = cumul_revenu - cumul_depense

            resultats.append({
                "Date": d.date(),
                "Revenu": cumul_revenu,
                "Depense": cumul_depense,
                "Ecart": cumul_ecart
            })

    else:
        if valeurs is None:
            valeurs = tab[colonne].unique()

        for v in valeurs:
            cumul_revenu = 0
            cumul_depense = 0
            cumul_ecart = 0

            for d in all_days:
                sous_tab = tab[(tab["Date"].dt.date == d.date()) & (tab[colonne] == v)]
                revenu = sous_tab[sous_tab["Type"] == "revenu"]["Valeur"].sum()
                depense = sous_tab[sous_tab["Type"] == "depense"]["Valeur"].sum()

                cumul_revenu += revenu
                cumul_depense += depense
                cumul_ecart = cumul_revenu - cumul_depense

                resultats.append({
                    "Date": d.date(),
                    colonne: v,
                    "Revenu": cumul_revenu,
                    "Depense": cumul_depense,
                    "Ecart": cumul_ecart
                })

    return pd.DataFrame(resultats).reset_index(drop=True)
```

```python
df_journalier = Cummuler_Journalier(Compte_Courant)
df_journalier
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# On suppose que df = Cummuler_Journalier(Compte_Courant)
df = Cummuler_Journalier(Compte_Courant)

fig, axes = plt.subplots(3, 1, figsize=(12, 8), sharex=True)

# Revenu
axes[0].plot(df["Date"], df["Revenu"], label="Revenu", marker="o")
axes[0].set_ylabel("Revenu")
axes[0].legend()
axes[0].grid(True)

# Dépense
axes[1].plot(df["Date"], df["Depense"], label="Dépense", color="red", marker="o")
axes[1].set_ylabel("Dépense")
axes[1].legend()
axes[1].grid(True)

# Ecart
axes[2].plot(df["Date"], df["Ecart"], label="Ecart", color="green", marker="o")
axes[2].set_ylabel("Ecart")
axes[2].legend()
axes[2].grid(True)

# Formateur de dates → mois abrégés (Jan, Fév, Mar, …)
month_fmt = mdates.DateFormatter("%b %Y")   # %b = mois abrégé, %Y = année
axes[2].xaxis.set_major_formatter(month_fmt)

# Ticks principaux = début de mois
axes[2].xaxis.set_major_locator(mdates.MonthLocator())

# Mise en forme
plt.xlabel("Mois")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# On suppose que df = Cummuler_Journalier(Compte_Courant)
df = Cummuler_Journalier(df_concat)
def plot_cumul(df):
    fig, axes = plt.subplots(2, 1, figsize=(12, 8), sharex=True)
    
    # Revenu
    axes[0].plot(df["Date"], df["Revenu"], label="Revenu", marker="o")
    axes[0].plot(df["Date"], df["Depense"], label="Dépense", color="red", marker="o")
    axes[0].set_ylabel("Evolution en Euros")
    axes[0].legend()
    axes[0].grid(True)
    
    
    # Ecart
    axes[1].plot(df["Date"], df["Ecart"], label="Ecart", color="green", marker="o")
    axes[1].set_ylabel("Ecart")
    axes[1].legend()
    axes[1].grid(True)
    
    # Formateur de dates → mois abrégés (Jan, Fév, Mar, …)
    month_fmt = mdates.DateFormatter("%b %Y")   # %b = mois abrégé, %Y = année
    axes[1].xaxis.set_major_formatter(month_fmt)
    
    # Ticks principaux = début de mois
    axes[1].xaxis.set_major_locator(mdates.MonthLocator())
    
    # Mise en forme
    plt.xlabel("Mois")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()



plot_cumul(df)
```
